```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## ***Importing Data***

```python
from google.colab import drive

#'https://drive.google.com/file/d/1I2CYwSv3nt9bTGsTqm-8Su9n-2sadG5q/view?usp=drive_link'
url = 'https://drive.google.com/file/d/1I2CYwSv3nt9bTGsTqm-8Su9n-2sadG5q/view?usp=drive_link'
file_id = url.split('/')[-2]
drive = 'https://drive.google.com/uc?id=' + file_id
data = pd.read_csv(drive)
data
```

```python
data.isna().sum()
```

### ***Data Pre-processing***

```python
data['Mortgage'] = data['Mortgage'].fillna(data['Mortgage'].mean())
data['Personal Loan'] = data['Personal Loan'].fillna(data['Personal Loan'].mean())
data['Securities Account'] = data['Securities Account'].fillna(data['Securities Account'].mean())

data.isna().sum()
```

```python
# Encoding categorical data

data = data.drop('ID',axis=1)

x = data.drop('CreditCard', axis =1)
y = data['CreditCard']

print("The independent variable X are",x)

print("The independent variable X are",y)
```

```python
x.head()
```

```python
y.head()
```

### ***Model Training***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4,random_state= None)

print("x train:\n",x_train)
print("x test:\n",x_test)
print("y train:\n",y_train)
print("y test:\n",y_test)
```

```python
from sklearn.preprocessing import StandardScaler
ssc = StandardScaler()
x_train = ssc.fit_transform(x_train)
x_test = ssc.transform(x_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf',random_state=None)
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.n_support_
```

```python
model.support_vectors_
```

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

Cm_train = confusion_matrix(y_train, y_pred_train)
print(Cm_train)

Cm_test = confusion_matrix(y_test, y_pred_test)
print(Cm_test)
```

### ***Accuracy Test***

```python
accuracy_train = accuracy_score(y_train, y_pred_train)
print("Accuracy:",(accuracy_train)*100, '%')
```

```python
accuracy_test = accuracy_score(y_test, y_pred_test)
print((accuracy_test)*100, '%')
```

```python
from sklearn.metrics import classification_report

# Additional performance metrics
report_train = classification_report(y_train, y_pred_train)
report_test = classification_report(y_test, y_pred_test)

print("Training Set Metrics:")
print(report_train)

print("Testing Set Metrics:")
print(report_test)

# Feature weights (coefficients) for linear kernel
if model.kernel == 'linear':
    feature_weights = model.coef_
    print("Feature Weights:")
    print(feature_weights)
```
